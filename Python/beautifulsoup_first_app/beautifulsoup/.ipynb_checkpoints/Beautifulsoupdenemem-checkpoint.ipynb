<h1 style="text-align:center"> BeautifulSoup Kütüphanesi ile Web Scapping İşlemleri </h1> 

Genellikler veri biliminde kullanılan bu kütüphane bir web sitesi üzerinden bizlere belirli koşullar altında araştırma yaparak bilgiler getirmeye olanak sağlamaktadır. Dinamik web projelerinde kullanılan ve belirli bir örüntüye göre tekrar eden sistemlerde belirli elementlerden verileri elde edebilmemizi ve <strong>Python</strong> dinamiklerini kullanarak elde edilen bu bilgilerden raporlama süreçelleri gerçekleştirmemize yaramaktadır.

```python
from bs4 import BeautifulSoup
```

Kütüphanenin import edilmesi için kullanılmıştır

```python
import requests
```

```python
r = requests.get("https://www.sikayetvar.com/akbank")
```

requests kodu request kütüphanesinin içerisinde yer alan ve o sayfaya ile bağlantı kurmamızı sağlayan bir kütüphanedir. get methodu ile bu şekilde isteğimizi yolluyoruz

```python
r
```

response olarak bizlere 200 dönmesi başarılı bir istek gerçekleştirdiğinizi gösterir. Bazı sayfalarda bu işlemleri botlar yaptığını düşünerek engelleme süreci gerçekleşir. Bunun için bir header (UserAgent) belirlemeli (kendi header iniz neyse o ile sisteme istek atmanız gerekmektedir)

User Agen bilginizi öğrenmek için : <a href="https://www.google.com/search?q=my+user+agent&oq=my+&aqs=chrome.1.69i57j35i39j69i59j0i512j46i512j69i60l3.4343j1j7&sourceid=chrome&ie=UTF-8" > User Agent Bilginiz </a> 

```python
user_agent = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36"}
```

şimdi bir daha istek atalım ve <bold>request.get()</bold> methodu içerisine bu parametremizi verelim

```python
r = requests.get("https://www.sikayetvar.com/akbank",headers=user_agent)
```

```python
r
```

* bu şekilde başarılı bir bağlantı kurulmuş bulunmaktadır.

* Bir sonraki aşamada ise features üzerinden istenilen şekilde derleme gerçekleştirilerek sayfanın kaynak kodunu soup kütüphanesi aracılığı ile sisteme dahil etmeye geldi

```python
soup = BeautifulSoup(r.content, features="html")
```

```python
soup
```

Görüldüğü üzere sayfanın tüm kaynak kodları elimize ulaştı

```python
soup.title
```

sayfanın başlığına bu şekilde ulaşabildik. Bir sonraki aşamada isteğim ise tek tek <strong>şikayetvar.com</strong> üzerinden ilk sayfadaki şikayet başlıklarını çekmek olacaktır soup tan gelen verileri find yöntemi ile bu şekilde çekebiliriz

```python
str1= soup.find_all("h2",attrs={"class": "complaint-title"})
```

```python
# bu method bizlere her bir secttion içerisindeki h2 lerin açıklamaları bizlere getirecektir
```

```python
for i in str1:
    print(i.text)
```

Görüldüğü üzere bu şekilde başlıklar bizlere gelmiş bulundu
